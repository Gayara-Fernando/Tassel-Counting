### Define a function of what should be done for each image


```python
# first get all the unique names of the files in files_list

all_names = []
for file in files_list:
    name = "_".join(file.split("_")[:-1])
    all_names.append(name)
```

```python
# how many unique names?
all_unique_names = np.unique(all_names)
```

```python
# how many unique files?
len(all_unique_names)
```

```python
# now for each np file in the folder, load the reduced images, and the density maps and get the sub-images and 
# corresponding counts

store_new_path = r"C:\Users\15739\Desktop\UNL - all\Summer 2022\Manuscript\Image_set_1\sub images and counts"

def get_subimage_and_count(file_name, kernel_size = 32, stride = 8):
    load_image = np.load(numpy_folder + "\\"+ file_name + "_resized.npy")
    load_densitymap = np.load(numpy_folder + "\\"+ file_name + "_densitymap.npy")
    
    img_height = load_image.shape[0]
    img_width = load_image.shape[1]
    
    density_sums = []
    catch_image_name_count = []
    counter = 0
    for i in  range(0, img_height, stride):
        for j in range(0, img_width, stride):
            sub_window = load_image[i: i + kernel_size, j : j + kernel_size,:]
            density = load_densitymap[i: i + kernel_size, j : j + kernel_size]
            counter = counter + 1
            dense_sum = np.sum(density)
            each_path = (store_new_path + "\\" + file_name + "_" + str(counter) + ".jpg")
            plt.imsave(each_path, sub_window)
            density_sums.append(dense_sum)
            catch_image_name_count.append([each_path, dense_sum])
    
    return(catch_image_name_count)
```

```python
returned_imageloc_counts = []
for file_name in all_unique_names:
    returned_stuff = get_subimage_and_count(file_name)
    returned_imageloc_counts.append(returned_stuff)   
```

```python
len(returned_imageloc_counts)
```

```python
all_data = []
for everything in returned_imageloc_counts:
    for data in everything:
        all_data.append(data)
```
